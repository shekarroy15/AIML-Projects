```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import boto3
from pandas import Series,DataFrame
from pandas_datareader import DataReader
from datetime import datetime
from __future__ import division
```

```python
def connecting_aws_s3():
    s3=boto3.resource(
            service_name='s3',
            region_name='us-east-1',
            aws_access_key_id='KEY',
            aws_secret_access_key='Secret_Key')
    print("connected")
    return s3
        
s3=connecting_aws_s3()
```

```python
def check_bucket(s3):
    for bucket in s3.buckets.all():
        print(bucket.name)

check_bucket(s3)
```

```python
def object_retrival(s3):
    for ob in s3.Bucket('testgl1').objects.all():
        print(ob)
        
object_retrival(s3)
```

```python
def read_files_s3(s3):
    ob=s3.Bucket('testgl1').Object('customer_data.csv').get()
    customer_df=pd.read_csv(ob['Body'])
    return customer_df

df=read_files_s3(s3)
df.head()
```
