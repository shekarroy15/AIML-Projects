```python
# Creating a Dataframe with all the information together

quote = []
author = []

for i in range(0, len(Quote)):
    quote.append(Quote[i].text)
    author.append(Author[i].text)

data = pd.DataFrame({"Quote": quote, "Author": author})
data.head()
```

```python
quote = []
author = []

for i in range(1, 11):
    headers = {
        'User-Agent':
        'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'
    }

    page = "http://quotes.toscrape.com/page/" + str(i)
    pageTree = requests.get(page, headers=headers)
    pageSoup = BeautifulSoup(pageTree.content, 'html.parser')

    Quote = pageSoup.find_all('span', {'class': 'text'})
    Author = pageSoup.find_all('small', {'class': 'author'})

    for i in range(0, len(Quote)):
        quote.append(Quote[i].text)
        author.append(Author[i].text)

    data = pd.DataFrame({"Quote": quote, "Author": author})
```
